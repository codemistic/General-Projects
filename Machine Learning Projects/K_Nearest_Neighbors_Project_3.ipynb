## Episode 1: Import Libraries
**Import pandas,seaborn, and the usual libraries.**

```python
#importing all required modules:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Episode 2: Get the Data
**Read the 'KNN_Data.csv' file into a dataframe**

```python
df = pd.read_csv("KNN_Data.csv")
```

Since these data are artificial, we will make a large pairplot with seaborn.

# Episode 4: Standardize the Variables

Time to standardize the variables.

**Import StandardScaler from Scikit learn.**

```python
#NOTE-It's always advisable to bring all features to the same scale for applying distance based algorithms like KNN or K-Means
"""because it will give higher weightage to variables which have higher magnitude and We do not want our algorithm to be affected
by the magnitude of these variables. The algorithm should not be biased towards variables with higher magnitude. 
To overcome this problem, we can bring down all the variables to the same scale and one of the most common technique to do
so is normalization where we calculate the mean and standard deviation of the variable."""

from sklearn.preprocessing import StandardScaler
```

**Create a StandardScaler() object called scaler.**

```python
scaler = StandardScaler()
```

**Fit scaler to the features.**

```python
scaler.fit(df.drop("TARGET_CLASS", axis=1))
```

**Use the .transform() method to convert the features to a scaled version.**

```python
scaled_features = scaler.transform(df.drop("TARGET_CLASS", axis=1))
```

**Convert the scaled features to a dataframe and check the head of this dataframe to make sure the scaling worked.**

```python
df_feat = pd.DataFrame(scaled_features, columns=df.columns[:-1])  #taking all columns accept the last one
df_feat.head()
```

# Episode 5: Train Test Split

**Use train_test_split to split your data into a training set and a testing set.**

```python
from sklearn.model_selection import train_test_split
```

```python
x=df_feat
y=df["TARGET_CLASS"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

# Episode 6: Using KNN

**Import KNeighborsClassifier from scikit learn.**

```python
from sklearn.neighbors import KNeighborsClassifier
```

**Create a KNN model instance with n_neighbors=1**

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

**Fit this KNN model to the training data.**

```python
knn.fit(x_train, y_train)
```

# Episode 7: Predictions and Evaluations
Let's evaluate our KNN model!

**Use the predict method to predict values using your KNN model and X_test.**

```python
preds = knn.predict(x_test)
preds
```

**Create a confusion matrix and classification report.**

```python
#importing evaluation matrices
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, preds))
```

```python
print(classification_report(y_test,preds))
```

# Episode 8: Choosing a K Value
Let's go ahead and use the elbow method to pick a good K Value!

**Create a for loop that trains various KNN models with different k values, then track the error_ rate for each of these models with a list. Check out the lesson if you are confused at this step.**



```python
#Elbow method to choose value of K
#Now, we wil use Elbow method to choose optimal value of k or n_neighbour parameter
#Elbow method is one of the popular methods to determine this optimal value of K

error_rate=[]

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean(pred_i!=y_test))  
```

**Now create the following plot using the information from your for loop.**

```python
plt.figure(figsize=(10,6))   #figure size =10*6
plt.plot(range(1,40),error_rate, color="blue", linestyle="dashed",marker="o", markerfacecolor="red", markersize=10)
plt.title("Error_rate vs K values")
plt.xlabel("K values")
plt.ylabel("Error_rate")
```

## Episode 9: Retrain with new K Value

**Retrain your model with the best K value (it's up to you to decide what you want) and re-do the classification report and the confusion matrix.**


```python
#According to the above graph, it is best to choose value of k approx 30 for minimizing the errors and increase accuracy
knn = KNeighborsClassifier(n_neighbors=30)
knn.fit(x_train, y_train)

preds = knn.predict(x_test)

print("With k=30 \n")
print("Confusion matrix= \n")
print(confusion_matrix(y_test, preds),"\n")
print("Classification Report= \n")
print(classification_report(y_test, preds))
```
